```python
# Calculate the number of instances per class 

import seaborn as sns

counts = pd.DataFrame(data["label"].value_counts())
counts["Digit"] = counts.index
counts.rename(columns={"label": "Number of Examples"}, inplace=True)
_ = sns.barplot(x="Digit", y="Number of Examples", data=counts)
```

A balanced dataset is crucial in avoiding biases for the class with the majority of instances.
